---
### ⚙️ Passo 0: Inicialização e Carga dos Dados
Vamos carregar a base de dados Iris diretamente do Scikit-Learn e convertê-la para um DataFrame do Pandas para facilitar a visualização.

```python
%pip install scikit-learn pandas numpy matplotlib
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Carregando o dataset Iris
iris = load_iris()

# Convertendo as features (X) para DataFrame
X = pd.DataFrame(iris.data, columns=iris.feature_names)

# Isolando o target (y)
y = iris.target

print("Base Iris carregada!")
print("Features (X) - Primeiro registros:")
display(X.head())
print("Classes do Target (y):", iris.target_names)
```

---
### 📐 Passo 1: Investigando as Dimensões dos Dados
Antes de dividir, vamos analisar a estrutura da tabela de entrada usando o atributo `.shape`.

```python
print("Dimensões de X (Features):", X.shape)
print("Dimensões de y (Target):", y.shape)

# Explique:
# O que representa o primeiro valor (linhas) e o segundo valor (colunas) da tupla acima?
```

---
### 🎯 Passo 2: Executando o Train/Test Split (70/30)
Dividiremos o dataset Iris na proporção de **70% para treino** e **30% para teste**, fixando a semente aleatória `random_state=42`.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.3, 
    random_state=42
)

print("--- Proporções da Divisão (70/30) ---")
print("Treino X (pesquisa/exercícios):", X_train.shape, "| Teste X (prova final):", X_test.shape)
print("Treino y (gabarito do treino):", y_train.shape, " | Teste y (gabarito da prova):", y_test.shape)
```

## 🧪 Aula 03 - Modelagem de Classificação: Árvores de Decisão

Agora que já carregamos os dados e realizamos a divisão em Treino e Teste (Passos 0, 1 e 2), iniciaremos o treinamento do nosso primeiro modelo preditivo: a **Árvore de Decisão**.

### 🌳 Passo 3: Treinando e Prevendo com Árvore de Decisão
Vamos instanciar a classe `DecisionTreeClassifier` definindo o limitador de profundidade máxima (`max_depth=3`) e treinar nos dados de treino.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

# 1. Instanciar o modelo de Árvore de Decisão (max_depth=3)
tree_model = DecisionTreeClassifier(max_depth=3, random_state=42)

# 2. Treinar o modelo nos dados de treino
tree_model.fit(X_train, y_train)

# 3. Gerar previsões para o conjunto de teste
y_pred_tree = tree_model.predict(X_test)

print("Modelo de Árvore de Decisão treinado com sucesso!")
```

---
### 🌳 Passo 4: Visualizando as Regras da Árvore
Execute a célula abaixo para visualizar graficamente as regras criadas pelo modelo.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 8))
plot_tree(
    tree_model,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,
    rounded=True
)
plt.title("Regras da Árvore de Decisão (max_depth=3)")
plt.show()
```
