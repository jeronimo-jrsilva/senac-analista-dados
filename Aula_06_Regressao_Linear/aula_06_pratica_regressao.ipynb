## Geração Dinâmica de Dados
Insira seu nome e data de nascimento abaixo para obter seu conjunto de dados de vendas exclusivo para a atividade.

```python
# -----------------------------------------------------
# CONFIGURAÇÃO DE INDIVIDUALIZAÇÃO DE DADOS
# Insira seu nome completo e data de nascimento abaixo!
# -----------------------------------------------------
NOME_ALUNO = "Seu Nome Aqui"
DATA_NASCIMENTO = "01012000"  # Formato DDMMAAAA

import numpy as np
import pandas as pd
import hashlib

# Gerador de semente determinística única por aluno
semente_str = f"{NOME_ALUNO.strip().lower()}_{DATA_NASCIMENTO}"
semente = int(hashlib.md5(semente_str.encode('utf-8')).hexdigest(), 16) % (2**32 - 1)
np.random.seed(semente)

# Geração de dados de Marketing e Vendas correlacionados
n_amostras = 150
investimento = np.random.uniform(500, 10000, n_amostras)  # R$ investidos em ads
# Cliques possuem correlação linear com investimento
cliques = (investimento * 0.15) + np.random.normal(50, 30, n_amostras)
# Vendas (Target) correlacionado com investimento e cliques + ruído
vendas = (investimento * 0.45) + (cliques * 1.2) + np.random.normal(1500, 500, n_amostras)

# Garantir valores positivos
cliques = np.clip(cliques, 10, None)
vendas = np.clip(vendas, 100, None)

# Criar DataFrame
df = pd.DataFrame({
    'Investimento_Marketing': np.round(investimento, 2),
    'Cliques_Anuncio': np.round(cliques).astype(int),
    'Vendas': np.round(vendas, 2)
})

print(f"🎉 Dataset individual gerado com sucesso para {NOME_ALUNO}!")
print(f"Semente gerada: {semente}")
df.head()
```
